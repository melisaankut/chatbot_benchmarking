```python
import pandas as pd
```

```python
# Read the .xlsx files
articles = pd.read_excel(r"data\articles.xlsx")
materials = pd.read_excel(r"data\materials.xlsx")
production_orders = pd.read_excel(r"data\production_orders.xlsx")
work_orders = pd.read_excel(r"data\work_orders.xlsx")
```

```python
# Group by drawn_no and count unique orderdescription values
anomalies = production_orders.groupby('drawno')['orderdesc'].nunique()

anomalies = anomalies[anomalies > 1]
anomaly_rows = production_orders[production_orders['drawno'].isin(anomalies.index)]


anomaly_rows.to_csv("anomalies.csv", index=False)
```

```python
# Group by materialid and count unique materialdescr values
anomalies = materials.groupby('materialid')['materialdescr'].nunique()

anomalies = anomalies[anomalies > 1]
anomaly_rows = materials[materials['materialid'].isin(anomalies.index)]


anomaly_rows.to_csv("anomalies.csv", index=False)
```

```python
# Group by materialname and count unique materialdescr values
anomalies = materials.groupby('materialname')['materialdescr'].nunique()

anomalies = anomalies[anomalies > 1]
anomaly_rows = materials[materials['materialname'].isin(anomalies.index)]


anomaly_rows.to_csv("anomalies.csv", index=False)
```

```python
# Group by productname and count unique description values
anomalies = articles.groupby('productname')['description'].nunique()

anomalies = anomalies[anomalies > 1]
anomaly_rows = articles[articles['productname'].isin(anomalies.index)]


anomaly_rows.to_csv("anomalies.csv", index=False)
```

```python
# 1. articles ve materials'ı orderid üzerinden birleştir
merged_orders_materials= pd.merge(production_orders, materials, on="orderid", how="outer", suffixes=('_article', '_material'))

merged_orders_materials.to_excel("merged_orders_materials.xlsx", index=False)
```
